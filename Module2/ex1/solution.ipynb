## Bài tập 3: Phân tích dữ liệu dạng bảng

```python
!gdown 1iA0WmVfW88HyJvTBSQDI5vesf-pgKabq
import pandas as pd

df = pd.read_csv ('./advertising.csv')

data = df.to_numpy()
```

```python
# 15. Find the max sale and index
max_sale_index = np.argmax(data[:, 3])

print(f'Max: {data[max_sale_index, 3]}, index: {max_sale_index}')
```

```python
# 16. Average of TV
average_tv = np.mean(data[:, 0])
print(f'Average TV: {average_tv}')
```

```python
# 17. Count the sales record >= 20
count_sales = np.sum(data[:, 3] >= 20)
print(f'Count sales >= 20: {count_sales}')
```

```python
# 18. Average of radio with sales >= 15
average_radio = np.mean(data[data[:, 3] >= 15, 1])
print(f'Average radio with sales >= 15: {average_radio}')
```

```python
# 19. sum of sales with newspaper > average newspaper
sum_sales = np.sum(data[data[:, 2] > np.mean(data[:, 2]), 3])
print(sum_sales)
```

```python
# 20. 
average_sales = np.mean(data[:, 3])

scores = np.where(data[:, 3] > average_sales, 'Good', np.where(data[:, 3] < average_sales, 'Bad', 'Average'))
print(scores[7:10])
```

```python
print(average_sales)
```

```python
# 21.
distance_average_sales = np.abs(data[:, 3] - average_sales)
neareast_average_sales = data[np.argmin(distance_average_sales), 3]

scores = np.where(data[:, 3] > neareast_average_sales, 'Good', np.where(data[:, 3] < neareast_average_sales, 'Bad', 'Average'))
print(scores[7:10])
```
